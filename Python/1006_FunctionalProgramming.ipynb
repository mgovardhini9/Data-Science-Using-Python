### List comprehensions :


#### Syntax :

[ele for iterVar in groupOfElements]

```python
li = [i for i in range(1, 101)]

print(li)
```

```python
li = [[i] for i in range(1, 101)]

print(li)
```

```python
li = ['*' for i in range(1, 101)]

print(li)
```

```python
li = [i for i in range(1, 101) if i % 2 == 0]

print(li)
```

```python
li = [year for year in range(2000, 3000) if year % 400 == 0 or (year % 4 == 0 and year % 100 != 0)]

print(li)
```

```python
li = ['even' if i % 2 == 0 else 'odd' for i in range(200, 300) ]

print(li)
```

### Dictionary Comprehensions :

#### create a dictionary key as a number and the value as the square of the number of 1 - 100

```python
di = {}

for num in range(1, 101) :
    di[num] = num ** 2
    
print(di)
```

```python
di = {i : i ** 2 for i in range(1, 101)}

print(di)
```

```python
di = {i : i+5 for i in range(1, 101)}

print(di)
```

#### key as the number and value as the square of previous number.

```python
di = {i : (i-1) ** 2 for i in range(1, 101)}

print(di)
```

```python
di = {i : i ** 2 for i in range(1, 101) if i % 2 == 0}

print(di)
```

```python
di = {i : 'Even' if i % 2 == 0 else 'odd' for i in range(1, 101)}

print(di)
```
